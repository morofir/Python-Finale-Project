```python
import pandas as pd
import numpy as np
```

# 1. Data Cleaning

```python
googleData=pd.read_csv(r'googleplaystore.csv',header=0)
googleData
```

```python
googleData=googleData.drop_duplicates()
len(googleData)
```

483 duplicates were removed

```python
googleData=googleData[np.isfinite(googleData['Rating'])]
len(googleData)
```

removing all lines where rating equals NaN

```python
googleData.dtypes
```

We see that except for Rating all columns are not numeric

Removing in:
    1. Installs the + and the ,
    2. Price the $
    3. in Size the , and the M and changing the Varies with device to NaN

```python
googleData['Installs'] = googleData['Installs'].str.replace('+','')
googleData['Installs'] = googleData['Installs'].str.replace(',','')
googleData['Price'] = googleData['Price'].str.replace('$','')
googleData['Size'] = googleData['Size'].str.replace(',','')
googleData['Size']=googleData['Size'].str.replace('M','')
googleData['Size']=googleData['Size'].replace('Varies with device',np.nan)

googleData.head(-100)
```

Reviews are converted to numeric values

```python
googleData['Reviews'] = googleData['Reviews'].apply(lambda x: float(str(x).replace('M', '')) * 1000000 if 'M' in str(x) else float(x))
googleData.nsmallest(10, 'Reviews')
googleData.nlargest(10, 'Reviews')
```

Dropping bad line

```python
googleData.drop([10472],axis=0,inplace=True)
```

Installs are converted to numeric values

```python
googleData['Installs']=pd.to_numeric(googleData['Installs'])
googleData.nsmallest(10, 'Installs')
googleData.nlargest(10,'Installs')
```

Price is converted to numeric values

```python
googleData['Price']=pd.to_numeric(googleData['Price'])
googleData.nsmallest(10, 'Price')
googleData.nlargest(10,'Price')
```

    Converting all Size cells that has KB to the proper ammount in MB
    Converting Size column to float
    Renaming Size column to SizeInMB

```python
googleData['Size'] = googleData['Size'].apply(lambda x: float(str(x).replace('k', '')) / 1000 if 'k' in str(x) else float(x))
googleData.rename(columns={'Size':'SizeInMB'},inplace=True)
googleData.nsmallest(10, 'SizeInMB')
googleData.nlargest(10,'SizeInMB')
```

```python
googleData.head(50)
```

Fill "Varies with device" with mean of size in each category for Size column

```python
googleData['SizeInMB'].fillna(googleData.groupby('Category')['SizeInMB'].transform('mean'),inplace=True)
googleData.head()
```
